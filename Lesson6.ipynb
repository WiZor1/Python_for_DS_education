```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
boston_keys = boston.feature_names
boston_trg = boston.target

X = pd.DataFrame(data=boston.data, columns=boston_keys)
y = pd.DataFrame(data=boston_trg, columns=['price'])

X_train,  X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
pred_train = lr.predict(X_train)
```

```python
r2_score(pred, y_test), r2_score(pred_train, y_train)
```

Сделайте предсказание на тестовых данных и посчитайте R2. Сравните с результатом из предыдущего задания.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
boston = load_boston()
boston_keys = boston.feature_names
boston_trg = boston.target

X = pd.DataFrame(data=boston.data, columns=boston_keys)
y = pd.DataFrame(data=boston_trg, columns=['price'])

X_train,  X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
model.fit(X_train, y_train.values[:, 0])
pred_rand_forest = model.predict(X_test)
pred_train_rand_forest = model.predict(X_train)
```

```python
r2_score(pred_rand_forest, y_test), r2_score(pred_train_rand_forest, y_train)
```

Наиболее точные результаты показал второй метод, однако, судя по отношению R2 предсказаний тестовых к обучаемым данным, второй метод с текущими параметрами более склонен к переобучению.

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('creditcard.csv')
# если я правильно понял и нормализацию нужно делать только для параметра дискретных значений Class:
df.value_counts('Class', normalize=True)
```

```python
pd.options.display.max_columns = 100
df[:10]
```

```python
y = df.Class
X = df.drop('Class', axis='columns')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.describe()
```

```python
parameters = [{
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7)
}]
```

```python
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
model.fit(X_train, y_train)
```

```python
best_par = model.best_params_
best_par
```

```python
y_pred_proba = model.predict_proba(X_test)[:, 1]
y_pred_proba[:10]
```

```python
roc_auc_score(y_true=y_test, y_score=y_pred_proba)
```
